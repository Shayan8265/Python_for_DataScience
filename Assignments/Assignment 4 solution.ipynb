```python
import json
from urllib.request import urlopen

with urlopen("https://api.github.com/search/repositories?q=created:>2023-01-01+language:python&sort=stars&order=desc&per_page=10") as response:
    source = response.read()

# print(source)

data = json.loads(source)
# print(len(data['items']))
# print(json.dumps(data, indent = 2))
```

```python
import pandas as pd
df = pd.DataFrame(data['items'])
for i in df:
    if i == 'owner':
        o = pd.Series(df[i])
    elif i == 'license':
        l = pd.Series(df[i])
    elif i == 'topics':
        t = pd.Series(df[i])
        
oo = pd.DataFrame(o)
ll =pd.DataFrame(l)
tt =pd.DataFrame(t)

# display(df)
display(oo,ll,tt)
```

```python
# Saving as csv
df.to_csv('Main_Table.csv')

# Saving as json
df = df.to_dict()
with open('Main_Table.json', 'w') as f:
    json.dump(df, f)
```

```python
oo = oo.to_dict()
with open('owner.json', 'w') as f:
    json.dump(oo, f)
ll = ll.to_dict()
with open('license.json', 'w') as f:
    json.dump(oo, f)
tt = tt.to_dict()
with open('topics.json', 'w') as f:
    json.dump(tt, f)
```
