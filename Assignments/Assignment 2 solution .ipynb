```python
Car = np.random.randint(100)
Jeep = np.random.randint(30)
Bike = np.random.randint(50)
Vespa = np.random.randint(10)

x_axis = ['car', 'jeep', 'bike', 'Vespa']
y_axis = [Car,Jeep,Bike,Vespa]

plt.bar(x_axis, y_axis)
plt.ylabel('Number of Sales')
plt.title('Frequency of Occurences')
```

```python
# Generate a histogram that shows the distribution of a numerical variable.
var = np.random.normal(size = 1000)
plt.hist(var, bins=20,alpha = 0.5)
```

```python
# Create a pie chart that shows the distribution of a categorical variable.
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
categories = ['Car', 'Jeep', 'Bike', 'Cycle', 'Vespa']
values = np.random.randint(1, 10, size=5)

# Create a pie chart
plt.pie(values, labels=categories)
plt.title('Distribution of Vehicle Sales')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values1 = np.random.randint(1, 10, size=5)
values2 = np.random.randint(1, 10, size=5)
values3 = np.random.randint(1, 10, size=5)

# stacked bar chart
plt.bar(categories, values1, label='Variable 1')
plt.bar(categories, values2, bottom=values1, label='Variable 2')
plt.bar(categories, values3, bottom=values1+values2, label='Variable 3')
plt.title('Comparison of multiple variables across different categories')
plt.xlabel('Categories')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Generate a box plot that shows the distribution of a numerical variable, including the median, quartiles, and outliers.
import numpy as np
import matplotlib.pyplot as plt
var = np.random.random(100)
print(var)
plt.boxplot(var)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# create a sample time series data
dates = pd.date_range('2023-01-01', '2023-02-06')
values = np.random.randint(1,100,len(dates))

plt.plot(dates, values)
plt.xticks(rotation = 90)
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series Plot')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# create some data for the plot
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

# set axis labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')

# display the plot
plt.show()
```
